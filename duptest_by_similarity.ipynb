```python
#한글(단어)만 추출
import re
hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
def kor(list):
    result=[]
    for i in list:
        result.append(hangul.sub('',i))
    return result
ka30=kor(a30['words'])

#숫자(유사도)만 추출
sutza = re.compile('\d+\.\d+')
def num(list):
    result=[]
    for i in list:
        result.append(float(sutza.findall(i)[0]))
    return result
na30=num(a30['words'])
```

```python
a30['words']=ka30
a30['sim']=na30
# del a30['Unnamed: 0']
# del a30['brand']
# del a30['words']
# del a30['emotion']
a30
```

```python
#카테고리 내 중복 제거
print(len(a30))
#브랜드간 독립(중복 무관)이므로 브랜드별로 split
a30b0 = a30[a30['brand']==0]
a30b1 = a30[a30['brand']==1]
a30b2 = a30[a30['brand']==2]
a30b3 = a30[a30['brand']==3]
a30b4 = a30[a30['brand']==4]
a30b5 = a30[a30['brand']==5]

# 브랜드별 단어 유사도순 정렬
a30b0=a30b0.sort_values(by=['sim'],ascending=False)
a30b1=a30b1.sort_values(by=['sim'],ascending=False)
a30b2=a30b2.sort_values(by=['sim'],ascending=False)
a30b3=a30b3.sort_values(by=['sim'],ascending=False)
a30b4=a30b4.sort_values(by=['sim'],ascending=False)
a30b5=a30b5.sort_values(by=['sim'],ascending=False)

#중복단어 제거 : 중복시 sim 높은 단어 생존
a30b0 = a30b0.drop_duplicates(['words','category'],keep='first')
a30b1 = a30b1.drop_duplicates(['words','category'],keep='first')
a30b2 = a30b2.drop_duplicates(['words','category'],keep='first')
a30b3 = a30b3.drop_duplicates(['words','category'],keep='first')
a30b4 = a30b4.drop_duplicates(['words','category'],keep='first')
a30b5 = a30b5.drop_duplicates(['words','category'],keep='first')

a30 = pd.concat([a30b0,
       a30b1,
       a30b2,
       a30b3,
       a30b4,
       a30b5]
      )
print(len(a30))
print(a30)

```

```python
### print('기존 키워드 개수 : ',len(a30[a30['sim']==1.0]))
# #유사도 0.7기준 필터링
# a30_07=a30[a30['sim']>=0.7]
# print('유사도 0.7이상 : ',len(a30_07))
# #유사도 0.6기준 필터링
# a30_06=a30[a30['sim']>=0.6]
# print('유사도 0.6이상 : ',len(a30_06))
# #유사도 0.5기준 필터링
# a30_05=a30[a30['sim']>=0.5]
# print('유사도 0.5이상 : ',len(a30_05))
# #유사도 0.4기준 필터링
# a30_04=a30[a30['sim']>=0.4]
# print('유사도 0.4이상 : ',len(a30_04))
# print(a30_04['words'].tolist())

def dup_cnt(dict):
    cnt2 = 0
    cnt3 = 0
    cnt4 = 0
    cnt5 = 0
    cnt6 = 0
    cnt7 = 0
    cnt8 = 0
    cnt9 = 0
    for k,v in dict.items():
        if v==2:            cnt2+=1
        if v==3:            cnt3+=1
        if v==4:            cnt4+=1
        if v==5:            cnt5+=1
        if v==6:            cnt6+=1
        if v==7:            cnt7+=1
        if v==8:            cnt8+=1
        if v==9:            cnt9+=1
    print('2개 카테고리 중복 단어 : ',cnt2)
    print('3개 카테고리 중복 단어 : ',cnt3)
    print('4개 카테고리 중복 단어 : ',cnt4)
    print('5개 카테고리 중복 단어 : ',cnt5)
    print('6개 카테고리 중복 단어 : ',cnt6)
    print('7개 카테고리 중복 단어 : ',cnt7)
    print('8개 카테고리 중복 단어 : ',cnt8)
    print('9개 카테고리 중복 단어 : ',cnt9)
    print()
    return

def inner_dup(list):
    dup_cnt = {}
    result = {}
    for i in list:
        try : dup_cnt[i] += 1
        except : dup_cnt[i] = 1
    for k,v in dup_cnt.items():
        if v!=1 :result[k]=v
    return result

def total_num(total_df):
    brandlist=[
        '전체',
        '네이버',
        '카카오',
        'SKT',
        'KT',
        'LG'
    ]
    print('기존 키워드 개수 : ',len(total_df[total_df['sim']==1.0].drop_duplicates(['words'])))
    
    for i in range(6):
        print('======================================')
        print('브랜드 : ',brandlist[i])
        df=total_df[total_df['brand']==i]
        
        #유사도 0.7기준 필터링
        df_07=df[df['sim']>=0.7]
        print('유사도 0.7이상 : ',len(df_07))    
        ls07 = df_07['words'].tolist()
        dup_cnt(inner_dup(ls07))
        
        #유사도 0.6기준 필터링
        df_06=df[df['sim']>=0.6]
        print('유사도 0.6이상 : ',len(df_06))
        ls06 = df_06['words'].tolist()
        dup_cnt(inner_dup(ls06))
        
        #유사도 0.5기준 필터링
        df_05=df[df['sim']>=0.5]
        print('유사도 0.5이상 : ',len(df_05))
        ls05 = df_05['words'].tolist()
        dup_cnt(inner_dup(ls05))
        
        #유사도 0.4기준 필터링
        df_04=df[df['sim']>=0.4]
        print('유사도 0.4이상 : ',len(df_04))
        ls04 = df_04['words'].tolist()
        dup_cnt(inner_dup(ls04))
    return 

total_num(a30)
# total_num(a30b1)
# total_num(a30b2)
# total_num(a30b3)
# total_num(a30b4)
# total_num(a30b5)
```
