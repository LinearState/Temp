```python
#카테고리 내 중복 검사
def inner_dup(list):
    dup_cnt = {}
    result = {}
    for i in list:
        try : dup_cnt[i] += 1
        except : dup_cnt[i] = 1
    for k,v in dup_cnt.items():
        if v!=1 :result[k]=v
    return result
inner_dup_a30c0 = inner_dup(a30c0)
inner_dup_a30c1 = inner_dup(a30c1)
inner_dup_a30c2 = inner_dup(a30c2)
inner_dup_a30c3 = inner_dup(a30c3)
inner_dup_a30c4 = inner_dup(a30c4)
inner_dup_a30c5 = inner_dup(a30c5)
inner_dup_a30c6 = inner_dup(a30c6)
inner_dup_a30c7 = inner_dup(a30c7)
inner_dup_a30c8 = inner_dup(a30c8)

inner_dup_a20c0 = inner_dup(a20c0)
inner_dup_a20c1 = inner_dup(a20c1)
inner_dup_a20c2 = inner_dup(a20c2)
inner_dup_a20c3 = inner_dup(a20c3)
inner_dup_a20c4 = inner_dup(a20c4)
inner_dup_a20c5 = inner_dup(a20c5)
inner_dup_a20c6 = inner_dup(a20c6)
inner_dup_a20c7 = inner_dup(a20c7)
inner_dup_a20c8 = inner_dup(a20c8)

inner_dup_a10c0 = inner_dup(a10c0)
inner_dup_a10c1 = inner_dup(a10c1)
inner_dup_a10c2 = inner_dup(a10c2)
inner_dup_a10c3 = inner_dup(a10c3)
inner_dup_a10c4 = inner_dup(a10c4)
inner_dup_a10c5 = inner_dup(a10c5)
inner_dup_a10c6 = inner_dup(a10c6)
inner_dup_a10c7 = inner_dup(a10c7)
inner_dup_a10c8 = inner_dup(a10c8)

import pprint
pprint.pprint(inner_dup_a10c0)
```

```python
def dup_cnt(list):
    cnt2=0
    cnt3=0
    cnt4=0
    cnt5=0
    cnt6=0
    cnt7=0
    cnt8=0
    cnt9=0
    for i in list:
#         print(i)
        v=i[1]
        if v==2:
            cnt2+=1
        if v==3:
            cnt3+=1
        if v==4:
            cnt4+=1
        if v==5:
            cnt5+=1
        if v==6:
            cnt6+=1
        if v==7:
            cnt7+=1
        if v==8:
            cnt8+=1
        if v==9:
            cnt9+=1
    
    print('2개 카테고리 중복 단어 : ',cnt2)
    print('3개 카테고리 중복 단어 : ',cnt3)
    print('4개 카테고리 중복 단어 : ',cnt4)
    print('5개 카테고리 중복 단어 : ',cnt5)
    print('6개 카테고리 중복 단어 : ',cnt6)
    print('7개 카테고리 중복 단어 : ',cnt7)
    print('8개 카테고리 중복 단어 : ',cnt8)
    print('9개 카테고리 중복 단어 : ',cnt9)
    print()
    return
print('word2vec 30개 확장 시 {0}개 단어 중 {1}단어 중복\n중복 없는 단어 : {2}'.format(len(uniq_a30),len(dup_a30),len(uniq_a30)-len(dup_a30)))
dup_cnt(dup_a30)
print('word2vec 20개 확장 시 {0}개 단어 중 {1}단어 중복\n중복 없는 단어 : {2}'.format(len(uniq_a20),len(dup_a20),len(uniq_a20)-len(dup_a20)))
dup_cnt(dup_a20)
print('word2vec 10개 확장 시 {0}개 단어 중 {1}단어 중복\n중복 없는 단어 : {2}'.format(len(uniq_a10),len(dup_a10),len(uniq_a10)-len(dup_a10)))
dup_cnt(dup_a10)
```
